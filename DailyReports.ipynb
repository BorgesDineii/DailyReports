```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders
import os

def buscando_preco_acao(ticker_symbol):
    ticker = yf.Ticker(ticker_symbol)
    preco_atual = ticker.history(period="1d")['Close'][0]
    return preco_atual

# Configurações do servidor SMTP
smtp_host = 'smtp.gmail.com'  # Exemplo: smtp.gmail.com
smtp_port = 587  # Porta 587 para TLS
smtp_user = 'email do usuario'
smtp_password = 'senha smtp do email'

preco_bancointer = buscando_preco_acao("INBR32.SA")
preco_vale = buscando_preco_acao("VALE")
# preco_leve3 = buscando_preco_acao("LEVE3.SA")
preco_positivo = buscando_preco_acao("POSI3.SA")
preco_bancopine = buscando_preco_acao("PINE4.SA")
preco_kepler = buscando_preco_acao("KEPL3.SA")
preco_gerdau = buscando_preco_acao("GGBR4.SA")
preco_banco_do_brasil = buscando_preco_acao("BBAS3.SA")
preco_zamp = buscando_preco_acao("ZAMP3.SA")
preco_b3 = buscando_preco_acao("B3SA3.SA")


"VALE", "INBR32.SA", "POSI3.SA", "PINE4.SA", "KEPL3.SA", "GGBR4.SA", "BBAS3.SA", "ZAMP3.SA", "B3SA3.SA"

# Configuração do e-mail
remetente = 'email do remetente'
destinatario = ['email']
assunto = 'Relatório de Mercado'
corpo_email = f'''Opa futuro milionário, Bão?

Segue relatório do mercado da data de hoje:


• O Banco Inter teve o retorno de {retorno_bancointer} a cotação esta de R${preco_bancointer:.2f}.

• As ações da Vale S.A tiveram um retorno de {retorno_vale} a cotação esta de R${preco_vale:.2f}.

• As ações da Positivo tiveram um retorno de {retorno_positivo} a cotação esta de R${preco_positivo:.2f}.

• As ações da Banco Pine tiveram um retorno de {retorno_banco_pine} a cotação esta de R${preco_bancopine:.2f}.

• As ações da Kepler Weber tiveram um retorno de {retorno_kepler_weber} a cotação esta de R${preco_kepler:.2f}.

• As ações da Gerdau tiveram um retorno de {retorno_gerdau} a cotação esta de R${preco_gerdau:.2f}.

• As ações da Banco do Brasil tiveram um retorno de {retorno_banco_do_brasil} a cotação esta de R${preco_banco_do_brasil:.2f}.

• As ações da Zamp tiveram um retorno de {retorno_zamp} a cotação esta de R${preco_zamp:.2f}.

• As ações da B3 tiveram um retorno de {retorno_b3} a cotação esta de R${preco_b3:.2f}.


Segue em anexo a performance dos ativos nos últimos 6 meses.

Atte,
Borges
'''

# Caminhos dos anexos
anexo_bancointer = "BancoInter.png"
anexo_vale = "Vale.png"
anexo_positivo = "positivo.png"
anexo_banco_pine = "banco_pine.png"
anexo_kepler_weber = "kepler_weber.png"
anexo_gerdau = "gerdau.png"
anexo_banco_do_brasil = "banco_do_brasil.png"
anexo_zamp = "zamp.png"
anexo_b3 = "b3.png"


# Compor o e-mail usando MIME
msg = MIMEMultipart()
msg['From'] = remetente
msg['To'] = ', '.join(destinatario)
msg['Subject'] = assunto
msg.attach(MIMEText(corpo_email, 'plain'))

# Função para anexar arquivos
def anexar_arquivo(msg, filepath):
    with open(filepath, 'rb') as f:
        part = MIMEBase('application', 'octet-stream')
        part.set_payload(f.read())
        encoders.encode_base64(part)
        part.add_header(
            'Content-Disposition',
            f'attachment; filename= {os.path.basename(filepath)}',
        )
        msg.attach(part)

# Anexar os arquivos
anexar_arquivo(msg, anexo_bancointer)
anexar_arquivo(msg, anexo_vale)
# anexar_arquivo(msg, anexo_leve3)
anexar_arquivo(msg, anexo_positivo)
anexar_arquivo(msg, anexo_banco_pine)
anexar_arquivo(msg, anexo_kepler_weber)
anexar_arquivo(msg, anexo_gerdau)
anexar_arquivo(msg, anexo_banco_do_brasil)
anexar_arquivo(msg, anexo_zamp)
anexar_arquivo(msg, anexo_b3)

# Enviar o e-mail
try:
    # Conectar ao servidor SMTP
    server = smtplib.SMTP(smtp_host, smtp_port)
    server.starttls()  # Iniciar a conexão TLS
    server.login(smtp_user, smtp_password)  # Fazer login

    # Enviar o e-mail
    server.sendmail(remetente, destinatario, msg.as_string())
    print("E-mail enviado com sucesso!")
except Exception as e:
    print(f"Erro ao enviar e-mail: {e}")
finally:
    server.quit()  # Fechar a conexão SMTP
```
